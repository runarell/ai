```python


# 필요 모듈 불러오기
import requests # 웹 페이지 연결 모듈    
from bs4 import BeautifulSoup as bs  # 읽어온 웹페잊를 분석하기 쉽게 정제
import pandas as pd 

url = "https://movie.naver.com/movie/sdb/rank/rmovie.naver";    # url 페이지 주소
html = requests.get( url )                               # 페이지 코드 긁어오기
soup = bs(html.text);                                    # 사용이 용의하도록 BeautifulSoup을 이용하여 변환하여 페이지 리스트에 저장

move_data = [];                                                          # 영화 순위를 정리할 배열                        
moves = soup.select("td.title");                 # 페이지중 영화 데이터 배열에 담기
print(moves)

for i in range(1, len(moves) ):                                           # 노래데이터 배열 반복 
    rank = i
    title = moves[i].select("div > a")[0].text.strip()                       # 영화재목 추출 strip()로 정리
    
    move_data.append([rank, title]);                                    # move_data 에 영화의 정보를 저장
    
df = pd.DataFrame(move_data, columns = ["순위","영화재목"]);           # song_data에 저장된 데이타 표로 변경
print(df)
# df.to_excel('bugs.xlsx', index=False)                                      # 엑셀 저장 
```

```python
# 스타벅스 분석
# pip install selenium
from selenium import webdriver                          
from bs4 import BeautifulSoup as bs                       
import pandas as pd 

driver = webdriver.Chrome("chromedriver.exe");    # 크롬 디바이스
url = "https://www.starbucks.co.kr/store/store_map.do";    # url 페이지 주소
driver.get(url);  
```

```python
area_btn = "#container > div > form > fieldset > div > section > article.find_store_cont > article > header.loca_search > h3 > a";
driver.find_element_by_css_selector(area_btn).click();

soul_btn ="#container > div > form > fieldset > div > section > article.find_store_cont > article > article:nth-child(4) > div.loca_step1 > div.loca_step1_cont > ul > li:nth-child(1) > a"
driver.find_element_by_css_selector(soul_btn).click();
```

```python
all_btn = "#mCSB_2_container > ul > li:nth-child(1) > a";
driver.find_element_by_css_selector(all_btn).click();
```

```python
html = driver.page_source;       # 전체 페이지 소스 읽어오기
soup = bs(html);                 # bs로  데이타 정제

starbucks_list = soup.select("ul.quickSearchResultBoxSidoGugun > li.quickResultLstCon"); 
st_list = []
for st in starbucks_list:
    name = st["data-name"]
    lat = st["data-lat"]
    lng = st["data-long"]
    adds = str(st.select("p")[0]).split("<br/>")[0].split(">")[1] 
    tel = str(st.select("p")[0]).split("<br/>")[1].split("<")[0] 
    
    st_list.append( [name, lat, lng, adds, tel] )

df = pd.DataFrame(st_list, columns = ["지점명","좌표1","좌표2", "주소", "전화번호"]);
print(df)

df.to_excel('str_adds.xlsx', index=False)   
```
