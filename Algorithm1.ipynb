```python
# 1. 알고리즘 이란?
# - 문제 과제를 해결하기 위한 처리 절파를 구체적인 순서로 표현한 아이디어 또는 생각
# - 알고리즘 다체는 일상생뢓에서도 많이 이미 사용하고 있다.
# - 요리레시피, 음악악보, 가전제품 사용설명서 등...
# - 문자, 기호, 일러스트 등으로 나타난다.

# 2. 알고리즘과 프로그램
# - 알고리즘을 프로그레밍 언어로 기술하면 프로그램이 된다.
# - 프로그래밍 언어(python, java, c ...)는 컴퓨터에 지시하기 위한 인공 언어

# 3. 프로그램 작성에서의 알고리즘
# - 프로그램 작성은 기획, 설계(*), 프로그램밍, 디버그 로 진행
# - 알고리즘은 설계 단계에서 필요하다. 

# 4. 좋은 알고리즘
# - 알기 쉽다. (쉬운 이해)
# - 속도가 빠르다.
# - 효율적이다. (메모리를 작게 사용한다.)
# - 재이용(범용성)이 쉽다.(객체지향 프로그래밍)

# 5. 왜 알고리즘을 공부해야 하나?
# - 좋은 프로그램을 만들기 위해! ( 빠르고, 효율적, 범용성)
# - 프로그램의 좋고 나쁨을 판단하기 위해
# - 프로그램 작성 과정 전체를 효율화하기 위해
# - 프로그램 기술을 향상시키기 위해

# 6. 알고리즘의 조건
# - 정확한 결과를 얻을 수 있어야 한다.
#   ( 알고리즘은 문제나 과제를 해결하기 위한 절차 )
# - 반듯이 종료되어야 한다.
#   ( 중간 처리나 절차가 잘못 만들어지면 영원히 끝나지 않는다. )

# 7. 알고리즘의 기본형 3가지
# - 순차 구조 : 처음부터 순서대로 처리하는 절차.
# - ex> 심부름 : 1.슈퍼마켓 간다. 2. 라면을 산다. 3. 집에 돌아온다.
#  실행을 원하는 처리를 위로부터 순서대로 작성한다.

# - 선택구조 : 조건식으로 판단해 실행할 절차를 전환하는 구조

# - 반복 구조 : 조건을 만조하는 동안 같은 처리를 반복하는 절차
    
```

```python
# 두 데이터의 대소르 판단하는 알고리즘
# 의사코드
# a, b를 숫자로 받는다
# a가 b보다 큰가?
# yes a가 크다 //no b가 크가

a = int(input("숫자1"));
b = int(input("숫자2"));
str = ""
if( a == b):
    str = "a와 b가 같다";
elif(a > b):
    str = "a가 크다";
else:
    str = "b가 크다"

print(str);
```
