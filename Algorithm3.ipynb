```python
# Binary Search 이진 탐색법
# 이진 탐색법은 탐색하는 범위를 절반씩 줄여서 탐색하는 법
# 데이터가 오름/내림차순으로 정렬되어 있어야 함


#이진 탐색 알고리즘
# 1. 가운데요소 선택
# 2. 가운데요소 비교
# 3. 탐색범위 절반


arr = [11, 13, 15, 17, 29, 31]   # 찾는 값 17

def bs(n, arr):
    head = 0;
    end = len(arr)-1;    
    while(True):
        cen = int( head+end/2 );
        if head == end:
            print("값이 없습니다.")
            return 0;
        elif arr[cen] == n:
            print(f"{cen}에서 탐색됨")
            return 1;    
        elif arr[cen] < n:
            head = cen-1;
        elif arr[cen] > n:
            end = cen+1;
    
print( bs(17, arr) );  
```

```python
# hesh 보관

def heshListIn( arr ):
    rang = int( len(arr)*1.5 ) +1
    arrD = [0 for i in range( len(rang) ];  # 데이타를 넣을 배열
    
    for i in arr:                           #  임시보관 데이터를 하나씩 반복
        k = i%len(arrD);                    # 데이터값 / 보관리스트 길이의 나머지

        for j in range(len(arrD))           # 보관배열의 길이만큼 반복하면 모든 칸을 확인한 것
            if arrD[1] != 0:                # 데이터 자리가 0이 아님 확인
                k = i%(k+1);                # i%k이면 나머지는 0~(k-1), (k-1)이상이면 0부터 반복
            else:
                arrD[k] = i;                # 데이타 보관
                break;


arrH = [12, 25, 36, 20, 30, 8, 42 ]; # 임시 보관 배열
arrD = heshListIn( arrH )

print(arrD)
print( len(arrD) )
```

```python
# 소수 판별
def isPrime (num):
    for i in range(2, math.floor(math.sqrt(num))+1):
        if num%i == 0 :
            return False
    return True

# 소수 찾기
def findPrime(n):
    primes = []
    for i in range(2, n+1):  # for i in range(2, (n//2)+1) 로 개선 가능
        if isPrime(i):
        primes.append(i)
    return primes

print('소수 리스트', findPrimes(30))

# 소인수 분해 1
def factorize(n):
  factors = []
  primes = findPrimes(n)  # n의 소수 리스트를 추출
  for i in primes:
      while (n % i == 0):  # 소수 중 나누어 떨어지는(약수) 수를 찾는다
          factors.append(i)
          n = n // i
  return factors
```
