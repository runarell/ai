```python
# Linear Search 선형탐색
# 보관된 데이터를 맨 앞부터 순서대로 원하는 값을 찾을 때까지 탐색
# 탐색 처리는 반복 구조를 사용하여 기술
# 반복 구조에서 종료 조건을 잊지 말자

arr = [1,2,3,4,6];
for i, num in enumerate(arr):
    if num == 5:
        print(f"{i}번째 요소가 일치");
        break;
    if num == arr[-1]:
        print("일치하는 요소가 없음");
           
```

```python
arr = [1,2,3,4,5];       

def liner_search(n, arr):
    st = "";
    for i, num in enumerate(arr):
        if num == n:
            st = f"{i}번째 요소가 일치";
            return st;
        
    return "일치하는 요소가 없음";
print(liner_search(5, arr));


```

```python
# Binary Search 이진 탐색법
# 이진 탐색법은 탐색하는 범위를 절반씩 줄여서 탐색하는 법
# 데이터가 오름/내림차순으로 정렬되어 있어야 함


#이진 탐색 알고리즘
# 1. 가운데요소 선택
# 2. 가운데요소 비교
# 3. 탐색범위 절반


arr = [11, 13, 23, 19, 17, 29, 31]   # 찾는 값 17
def bs(n, arr):
    ar = arr;
    while(len(ar) <= 1):
        cent = int(len(ar)/2)
        if ar[cent] == n:
            print("탐색됨")
            return 1;
        elif ar[cent] < n:
            ar = arr[cent-1:];
        elif ar[cent] > n:
            ar = arr[:cent+1];
        print( len(ar) );
    
print( bs(17, arr) );    
```
