```python
# 스타벅스 분석
# pip install selenium
from selenium import webdriver                          
from bs4 import BeautifulSoup as bs                       
import pandas as pd 

driver = webdriver.Chrome("chromedriver.exe");    # 크롬 디바이스
url = "https://www.starbucks.co.kr/store/store_map.do";    # url 페이지 주소
driver.get(url);  
```

```python
area_btn = "#container > div > form > fieldset > div > section > article.find_store_cont > article > header.loca_search > h3 > a";
driver.find_element_by_css_selector(area_btn).click();

soul_btn ="#container > div > form > fieldset > div > section > article.find_store_cont > article > article:nth-child(4) > div.loca_step1 > div.loca_step1_cont > ul > li:nth-child(1) > a"
driver.find_element_by_css_selector(soul_btn).click();
```

```python
all_btn = "#mCSB_2_container > ul > li:nth-child(1) > a";
driver.find_element_by_css_selector(all_btn).click();
```

```python
html = driver.page_source;       # 전체 페이지 소스 읽어오기
soup = bs(html);                 # bs로  데이타 정제

starbucks_list = soup.select("ul.quickSearchResultBoxSidoGugun > li.quickResultLstCon"); 
st_list = []
for st in starbucks_list:
    name = st["data-name"]
    lat = st["data-lat"]
    lng = st["data-long"]
    tel = st.select("p")[0].get_text(" ").split()[-1]
    adds = st.select("p")[0].get_text().strip(tel)
#     adds = str(st.select("p")[0]).split("<br/>")[0].split(">")[1] 
#     tel = str(st.select("p")[0]).split("<br/>")[1].split("<")[0] 
    
    st_list.append( [name, lat, lng, adds, tel] )

seoul_starbucks = pd.DataFrame(st_list, columns = ["지점명","좌표1","좌표2", "주소", "전화번호"]);
print(seoul_starbucks)

seoul_starbucks.to_excel("seoul_starbucks.xlsx", index=False)   
```

```python
# seoul_starbucks 데이터 프레임에 '시군구' 이름의 열을 만들어서
# 주소에서 구만 추출하여 삽입하시오.
sgg_names = []

for addr in seoul_starbucks['주소']:
    sgg = addr.split()[1]
    sgg_names.append(sgg)
    
seoul_starbucks['시군구명'] = sgg_names

seoul_starbucks.head()
```

```python
starbucks_count = seoul_starbucks.pivot_table(
                                             index = '시군구명',
                                             values='지점명',
                                             aggfunc='count')
starbucks_count
```
