```python
# method 매서드
# 클래스가 가지고 있는 함수를 매서드라고 부른다. 
# 매서드는 첫번째 매개변수로 self를 넣어야 한다. 


class Student:
  # 생성자 constructor
  def __init__(self, name, korean, math, eng, sci):
    self.name = name
    self.korean = korean
    self.math = math
    self.eng  = eng 
    self.sci = sci

  def get_sum(self):
    return self.korean + self.math + self.eng + self.sci

  def get_avg(self):
    return self.get_sum() / 4

  def to_string(self):
    return '{}\t\t{}\t\t{}'.format(self.name, self.get_sum(), self.get_avg())


# 학생 리스트

students = [
            Student('슈가', 87, 98, 88, 95),
            Student('진', 90, 99, 86, 97),
            Student('카리', 99, 89, 99, 99)
]

print('이름','총점','평균', sep='\t\t')
for student in students:
  print(student.to_string())

# 어떤 객체가 가지고 있는 어떤 함수(기능)을 명확하게 이해할 수 있으므로 매우 편리하다.

```

```python
result1 = 0;
result2 = 0;

def add1(num):
    global result1;
    result1 += num;
    return result1;

def add2(num):
    global result2;
    result2 += num;
    return result2;

print(add1(3))
print(add2(5))

print(add1(2))
print(add2(4))
```

```python
class Cal:
    # 생성자
    def __init__(self):
        self.result = 0;
    
    def add(self,num):
        self.result += num;
        return self.result;
    
    def sub(self,num):
        self.result -= num;
        return self.result;

cal1 = Cal();
cal2 = Cal();

print(cal1.add(3));
print(cal1.add(4));

print(cal2.add(5));
print(cal2.add(6));
```

```python
class FourCal:
    def setdata(self, first, second):
        self.first = first;
        self.second = second;
    def add(self):
        result = self.first + self.second;
        return result;
    
a = FourCal();
a.setdata(4,2);
a.add()

```

```python
# 상속 inheritance 
# 어떤 클래스를 만들때 다른 클래스의 기능을 물려받아 사용한다.

class MoreFourCal(FourCal):
    pass

a = MoreFourCal(4,2);
a.add()
```
