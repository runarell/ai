```python
# 집합 계산

set1 = {1,2,3};
set2= {2,3,4};

print(type(set1));
print(type(set2));

# 양쪽 집합에 모두 공통적으로 포함되는 원소 - > 교집합
print(set1 & set2);
print( set1.intersection(set2) );

# 양쪽 집합의 모든 원소 - > 합집합
print( set1 | set2 );
print(set1.union(set2));

# 두개의 집합의 차  - > 차집합
print( set1 - set2 );
print(set1.difference(set2));
print( set2 - set1 );
```

```python
# 입력) 당신의 점수를 입력하세요 >>> 95
# 출력) 당신의 점수는 95점이고 학점은 A학점이니다

point = int( input('당신의 성적을 입력하세요 : ') );
lank ='';
if point >= 90:
  lank = 'A'
elif point >= 80:
   lank = 'B'
elif point >= 70:
   lank = 'C'
elif point >= 60:
   lank = 'D'
else:
  lank = 'F'

print(f'당신의 점수는{point} 학점은 {lank}입니다.' );
```
