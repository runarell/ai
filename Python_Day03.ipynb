```python
score = int( input('당신의 점수를 입력하세요>>>'));
PM = int( score%10 );
lank2 ='';

if score >= 90:
  lank2 = 'A';
elif score >= 80:
  lank2 = 'B';
elif score >= 70:
  lank2 = 'C';
elif score >= 60:
  lank2 = 'D';
else:
  lank2 = 'F';

if lank2 == 'F':
  lank1 = '';
elif PM > 6:
  lank1 = '+';
elif PM > 3:
  lank1 = '0';
else:
  lank1 = '-';

print(f'당신의 점수는 {score}점이고 학점은 {lank2 + lank1}학점입니다.')
```
